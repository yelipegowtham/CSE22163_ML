```python
import numpy as np

# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative of sigmoid function for backpropagation
def sigmoid_derivative(x):
    return x * (1 - x)

# Initialize network parameters
input_layer_size = 2  # A and B inputs
hidden_layer_size = 2  # Two hidden nodes
output_layer_size = 1  # Single output node

np.random.seed(42)  # Seed for reproducibility

# Weights and biases initialization
V = np.random.uniform(low=-1, high=1, size=(input_layer_size, hidden_layer_size))
W = np.random.uniform(low=-1, high=1, size=(hidden_layer_size, output_layer_size))
b_hidden = np.random.uniform(low=-1, high=1, size=(1, hidden_layer_size))
b_output = np.random.uniform(low=-1, high=1, size=(1, output_layer_size))

# Training data for AND gate
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [0], [0], [1]])

# Training parameters
learning_rate = 0.05
epochs = 10000
error_threshold = 0.002

# Training the network
for epoch in range(epochs):
    # Forward propagation
    hidden_input = np.dot(X, V) + b_hidden
    hidden_output = sigmoid(hidden_input)
    final_input = np.dot(hidden_output, W) + b_output
    final_output = sigmoid(final_input)

    # Compute error
    error = y - final_output
    loss = np.mean(np.square(error))
    
    if loss <= error_threshold:
        print(f"Convergence reached at epoch {epoch}")
        break

    # Backpropagation
    d_final_output = error * sigmoid_derivative(final_output)
    error_hidden_layer = d_final_output.dot(W.T)
    d_hidden_output = error_hidden_layer * sigmoid_derivative(hidden_output)

    # Update weights and biases
    W += hidden_output.T.dot(d_final_output) * learning_rate
    b_output += np.sum(d_final_output, axis=0, keepdims=True) * learning_rate
    V += X.T.dot(d_hidden_output) * learning_rate
    b_hidden += np.sum(d_hidden_output, axis=0, keepdims=True) * learning_rate

# Output final trained parameters and results
print("Final hidden weights (V):", V)
print("Final output weights (W):", W)
print("Final hidden biases:", b_hidden)
print("Final output biases:", b_output)
print("Output from neural network after training:", final_output)
```

```python
import numpy as np

class NeuralNetwork:
    def __init__(self, input_size, hidden_size, output_size, learning_rate):
        self.weights_input_hidden = np.random.randn(input_size, hidden_size)
        self.weights_hidden_output = np.random.randn(hidden_size, output_size)
        self.learning_rate = learning_rate

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def sigmoid_derivative(self, x):
        return x * (1 - x)

    def forward(self, inputs):
        self.hidden = self.sigmoid(np.dot(inputs, self.weights_input_hidden))
        self.output = self.sigmoid(np.dot(self.hidden, self.weights_hidden_output))
        return self.output

    def backward(self, inputs, targets):
        output_error = targets - self.output
        output_delta = output_error * self.sigmoid_derivative(self.output)

        hidden_error = output_delta.dot(self.weights_hidden_output.T)
        hidden_delta = hidden_error * self.sigmoid_derivative(self.hidden)

        self.weights_hidden_output += self.hidden.T.dot(output_delta) * self.learning_rate
        self.weights_input_hidden += inputs.T.dot(hidden_delta) * self.learning_rate

    def train(self, inputs, targets, epochs=1000):
        for _ in range(epochs):
            self.forward(inputs)
            self.backward(inputs, targets)

# Example Inputs and Targets for AND Gate
inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
targets = np.array([[0], [0], [0], [1]])  # Reshaped to be a column vector

# Initialize and Train Neural Network
nn = NeuralNetwork(input_size=2, hidden_size=2, output_size=1, learning_rate=0.05)
nn.train(inputs, targets)

# Print the output after training
output = nn.forward(inputs)
print("Final Output after Training:")
print(output)

```
